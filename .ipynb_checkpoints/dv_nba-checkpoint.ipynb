```python
def resumen_exploratorio(df, nombre_df="df"):
    # Total de filas (para porcentaje de nulos)
    total_filas = len(df)

    resumen = pd.DataFrame({
        'columna': df.columns,
        'dtype': df.dtypes.astype(str),
        'n_nulos': df.isna().sum().values,
        '%_nulos': (df.isna().sum() / total_filas * 100).round(2).values,
        'n_valores_unicos': df.nunique(dropna=True).values
    })

    # Ordenamos por % de nulos descendente porque suele ser lo primero a revisar
    resumen = resumen.sort_values(by='%_nulos', ascending=False).reset_index(drop=True)

    print(f"--- Resumen de {nombre_df} ---")
    print(f"Filas: {df.shape[0]} | Columnas: {df.shape[1]}")
    return resumen
```

```python
def top_values_por_columna(df, num_vals=5):
    resultado = {}

    for col in df.columns:
        vc = (
            df[col]
            .value_counts(dropna=False)
            .head(num_vals)
        )
        resultado[col] = vc

    return resultado
```

```python
def iqr_outlier_counts(df: pd.DataFrame, k: float = 1.5) -> pd.Series:
    """
    Retorna un Series con el conteo de outliers por columna numérica usando la regla de k*IQR.
    k=1.5 es el valor clásico; puedes usar 3.0 para detectar solo outliers extremos.
    """
    num = df.select_dtypes(include="number").replace([np.inf, -np.inf], np.nan)
    counts = {}
    for col in num.columns:
        s = num[col].dropna()
        if s.empty:
            counts[col] = 0
            continue
        q1 = s.quantile(0.25)
        q3 = s.quantile(0.75)
        iqr = q3 - q1
        if iqr == 0:  # columna constante → sin outliers por IQR
            counts[col] = 0
            continue
        lower = q1 - k * iqr
        upper = q3 + k * iqr
        counts[col] = (s.lt(lower) | s.gt(upper)).sum()
    return pd.Series(counts, name="outliers").sort_values(ascending=False)
```

```python
def boxplots_for_columns(
    df: pd.DataFrame,
    columns: list,
    k: float = 1.5,
    orient: str = "v",          # "v" (vertical) o "h" (horizontal)
    cols_per_row: int = 3,
    figsize: tuple | None = None,
    show_thresholds: bool = True,
    share_axis: bool = False,
):
    """
    Dibuja box plots para las columnas indicadas.
    - k: multiplicador del IQR (1.5 clásico; 3.0 para outliers extremos)
    - orient: "v" vertical o "h" horizontal
    - cols_per_row: cuántas gráficas por fila
    - show_thresholds: dibuja líneas en los umbrales Q1−k*IQR y Q3+k*IQR
    - share_axis: comparte eje entre subplots (útil cuando tienen escala similar)
    """
    # Filtra solo columnas numéricas manteniendo el orden dado
    num_cols = []
    skipped = []
    for c in columns:
        if pd.api.types.is_numeric_dtype(df[c]):
            num_cols.append(c)
        else:
            skipped.append(c)

    if not num_cols:
        raise ValueError("No hay columnas numéricas en la lista proporcionada.")

    n = len(num_cols)
    rows = ceil(n / cols_per_row)
    cols = min(cols_per_row, n)

    if figsize is None:
        figsize = (cols * 5, rows * 4)

    fig, axes = plt.subplots(rows, cols, figsize=figsize, sharey=(share_axis and orient=="v"), sharex=(share_axis and orient=="h"))
    if n == 1:
        axes = np.array([axes])
    axes = axes.flatten()

    for i, col in enumerate(num_cols):
        ax = axes[i]
        s = df[col].replace([np.inf, -np.inf], np.nan).dropna()

        # Calcula IQR y outliers para el título/umbrales
        if len(s) >= 1:
            q1 = s.quantile(0.25)
            q3 = s.quantile(0.75)
            iqr = q3 - q1
            if iqr == 0:
                lower = q1
                upper = q3
                outliers_count = 0
            else:
                lower = q1 - k * iqr
                upper = q3 + k * iqr
                outliers_count = (s.lt(lower) | s.gt(upper)).sum()
        else:
            lower = upper = np.nan
            outliers_count = 0

        # Dibuja el boxplot (whis controla los bigotes con k*IQR)
        ax.boxplot(
            [s.values] if orient == "v" else s.values,
            vert=(orient == "v"),
            whis=k,
            showfliers=True,
            patch_artist=False
        )

        # Etiquetas y título
        ax.set_title(f"{col}  |  n={len(s)}  |  outliers={outliers_count}")
        if orient == "v":
            ax.set_xticks([])
            ax.set_xlabel(col)
            ax.set_ylabel(col)
            if show_thresholds and len(s) > 0 and not np.isnan(lower):
                ax.axhline(lower, linestyle="--", linewidth=1)
                ax.axhline(upper, linestyle="--", linewidth=1)
        else:
            ax.set_yticks([])
            ax.set_ylabel(col)
            ax.set_xlabel(col)
            if show_thresholds and len(s) > 0 and not np.isnan(lower):
                ax.axvline(lower, linestyle="--", linewidth=1)
                ax.axvline(upper, linestyle="--", linewidth=1)

    # Oculta ejes sobrantes
    for j in range(i + 1, len(axes)):
        axes[j].axis("off")

    fig.suptitle(f"Box plots (IQR, k={k})", y=0.98)
    fig.tight_layout()
    return fig, axes
```

```python
def plot_team_wins(df, kind='top', n_teams=10):
    """
    Grafica los equipos con más o menos victorias.

    Parámetros
    ----------
    df : pandas.DataFrame
        DataFrame que contiene la columna 'winner_str'.
    kind : str, opcional
        'top'  -> muestra los equipos con más victorias
        'bottom' -> muestra los equipos con menos victorias
    n_teams : int, opcional
        Número de equipos a visualizar.
    """
    # Conteo de victorias por equipo
    wins = df['winner_str'].value_counts().reset_index()
    wins.columns = ['team', 'wins']

    # Seleccionar top o bottom
    if kind == 'top':
        wins_plot = wins.sort_values('wins', ascending=False).head(n_teams)
        title = f"Top {n_teams} equipos con más victorias"
        ascending_order = False
    elif kind == 'bottom':
        wins_plot = wins.sort_values('wins', ascending=True).head(n_teams)
        title = f"Bottom {n_teams} equipos con menos victorias"
        ascending_order = True
    else:
        raise ValueError("El parámetro 'kind' debe ser 'top' o 'bottom'.")

    plt.figure(figsize=(10, 6))
    sns.barplot(
        data=wins_plot,
        y='team',
        x='wins',
        order=wins_plot.sort_values('wins', ascending=ascending_order)['team']
    )
    plt.title(title)
    plt.xlabel("Número de victorias")
    plt.ylabel("Equipo")
    plt.tight_layout()
    plt.show()
```

```python
def plot_wins_by_year(df, kind='top', n_teams=3):
    """
    Grafica la evolución de victorias por año para los equipos más o menos ganadores.

    Parámetros
    ----------
    df : pandas.DataFrame
        DataFrame que contiene las columnas 'year' y 'winner_str'.
    kind : str, opcional
        'top'    -> equipos con más victorias totales
        'bottom' -> equipos con menos victorias totales
    n_teams : int, opcional
        Número de equipos a visualizar.
    """

    # Victorias por año y equipo
    wins_by_year = (
        df.groupby(['year', 'winner_str'])
          .size()
          .reset_index(name='wins')
    )

    # Victorias totales por equipo (sumando todos los años)
    total_wins = (
        wins_by_year.groupby('winner_str')['wins']
                    .sum()
                    .reset_index(name='total_wins')
    )

    # Seleccionar top o bottom
    if kind == 'top':
        selected_teams = (
            total_wins.sort_values('total_wins', ascending=False)
                      .head(n_teams)['winner_str']
        )
        title_kind = "más victorias"
    elif kind == 'bottom':
        selected_teams = (
            total_wins.sort_values('total_wins', ascending=True)
                      .head(n_teams)['winner_str']
        )
        title_kind = "menos victorias"
    else:
        raise ValueError("El parámetro 'kind' debe ser 'top' o 'bottom'.")

    # Filtrar solo esos equipos
    wins_by_year_sel = wins_by_year[wins_by_year['winner_str'].isin(selected_teams)]

    plt.figure(figsize=(12, 6))
    sns.lineplot(
        data=wins_by_year_sel,
        x='year',
        y='wins',
        hue='winner_str',
        marker='o'
    )
    plt.title(f"Evolución de victorias por año ({title_kind}, Top {n_teams} equipos)")
    plt.xlabel("Año")
    plt.ylabel("Número de victorias")
    plt.tight_layout()
    plt.show()
```

```python
def plot_home_away_wins(df, n_teams=10):
    """
    Grafica victorias como local vs visitante para los equipos con más victorias totales.

    Parámetros
    ----------
    df : pandas.DataFrame
        DataFrame con las columnas 'winner_str', 'hometeamName', 'awayteamName'.
    n_teams : int, opcional
        Número de equipos a visualizar (por defecto 10).
    """

    # Victorias como local (nombre del ganador coincide con hometeamName)
    home_wins = (
        df[df['winner_str'] == df['hometeamName']]
        .groupby('winner_str')
        .size()
        .rename('home_wins')
    )

    # Victorias como visitante
    away_wins = (
        df[df['winner_str'] == df['awayteamName']]
        .groupby('winner_str')
        .size()
        .rename('away_wins')
    )

    home_away = (
        pd.concat([home_wins, away_wins], axis=1)
          .fillna(0)
          .reset_index()
          .rename(columns={'winner_str': 'team'})
    )

    home_away['total_wins'] = home_away['home_wins'] + home_away['away_wins']
    home_away_top = home_away.sort_values('total_wins', ascending=False).head(n_teams)

    home_away_melt = home_away_top.melt(
        id_vars='team',
        value_vars=['home_wins', 'away_wins'],
        var_name='condition',
        value_name='wins'
    )

    plt.figure(figsize=(10, 6))
    sns.barplot(
        data=home_away_melt,
        x='team',
        y='wins',
        hue='condition'
    )
    plt.title(f"Victorias como local vs visitante (Top {n_teams} equipos)")
    plt.xlabel("Equipo")
    plt.ylabel("Número de victorias")
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
```

```python
def plot_win_rate(df, kind='top', n_teams=10, min_games=20):
    """
    Grafica el porcentaje de victorias (win rate) por equipo.

    Parámetros
    ----------
    df : pandas.DataFrame
        DataFrame con las columnas:
        'hometeamName', 'awayteamName', 'winner_str'.
    kind : str, opcional
        'top'    -> equipos con mayor win rate
        'bottom' -> equipos con menor win rate
    n_teams : int, opcional
        Número de equipos a visualizar.
    min_games : int, opcional
        Mínimo de partidos jugados para incluir al equipo.
    """

    # Juegos como local / visitante
    home_games = df.groupby('hometeamName').size().rename('home_games')
    away_games = df.groupby('awayteamName').size().rename('away_games')

    games = pd.concat([home_games, away_games], axis=1).fillna(0)
    games['total_games'] = games['home_games'] + games['away_games']

    # Victorias por equipo (usando winner_str)
    wins_team = df.groupby('winner_str').size().rename('wins')

    stats_team = pd.concat([games, wins_team], axis=1).fillna(0)
    stats_team['win_rate'] = stats_team['wins'] / stats_team['total_games']

    stats_team = stats_team.reset_index().rename(columns={'index': 'team'})

    # Filtro por mínimo de partidos
    stats_team_filt = stats_team[stats_team['total_games'] >= min_games]

    # Seleccionar top o bottom según win_rate
    if kind == 'top':
        stats_team_filt = stats_team_filt.sort_values('win_rate', ascending=False).head(n_teams)
        title_kind = "mayor win rate"
        ascending_order = False
    elif kind == 'bottom':
        stats_team_filt = stats_team_filt.sort_values('win_rate', ascending=True).head(n_teams)
        title_kind = "menor win rate"
        ascending_order = True
    else:
        raise ValueError("El parámetro 'kind' debe ser 'top' o 'bottom'.")

    plt.figure(figsize=(10, 6))
    sns.barplot(
        data=stats_team_filt,
        y='team',
        x='win_rate',
        order=stats_team_filt.sort_values('win_rate', ascending=ascending_order)['team']
    )
    plt.title(f"Porcentaje de victorias por equipo ({title_kind}, Top {n_teams})")
    plt.xlabel("Win rate")
    plt.ylabel("Equipo")
    plt.tight_layout()
    plt.show()
```

```python
def plot_home_win_heatmap(df):
    """
    Grafica un mapa de calor con el número de victorias del equipo local
    frente a cada equipo visitante.

    Parámetros
    ----------
    df : pandas.DataFrame
        DataFrame con las columnas 'winner_str', 'hometeamName', 'awayteamName'.
    """

    # Partidos donde gana el equipo local
    df_home_wins = df[df['winner_str'] == df['hometeamName']]

    # Matriz: filas = local, columnas = visitante, valores = victorias del local
    matrix_matchup = pd.crosstab(
        df_home_wins['hometeamName'],
        df_home_wins['awayteamName']
    )

    plt.figure(figsize=(12, 10))
    sns.heatmap(
        matrix_matchup,
        cmap="viridis",
        linewidths=0.5
    )
    plt.title("Mapa de calor: victorias del local frente al visitante")
    plt.xlabel("Equipo visitante")
    plt.ylabel("Equipo local")
    plt.tight_layout()
    plt.show()
```

```python
def plot_scorers(player_stats, kind='top', n_players=10, per='per_game', min_games=20):
    """
    Gráfica jugadores que más / menos anotan.

    per: 'per_game' o 'per_20'
    """
    if per == 'per_game':
        metric = 'points_per_game'
        metric_label = 'Puntos por partido'
    elif per == 'per_20':
        metric = 'points_per_20'
        metric_label = 'Puntos por 20 minutos'
    else:
        raise ValueError("per debe ser 'per_game' o 'per_20'.")

    df_plot = player_stats[player_stats['games_played'] >= min_games].copy()

    if kind == 'top':
        df_plot = df_plot.sort_values(metric, ascending=False).head(n_players)
        title_kind = 'más anotadores'
        asc = False
    elif kind == 'bottom':
        df_plot = df_plot.sort_values(metric, ascending=True).head(n_players)
        title_kind = 'menos anotadores'
        asc = True
    else:
        raise ValueError("kind debe ser 'top' o 'bottom'.")

    plt.figure(figsize=(10, 6))
    sns.barplot(
        data=df_plot,
        y='fullName',
        x=metric,
        order=df_plot.sort_values(metric, ascending=asc)['fullName']
    )
    plt.title(f"Jugadores {title_kind} ({metric_label})")
    plt.xlabel(metric_label)
    plt.ylabel("Jugador")
    plt.tight_layout()
    plt.show()
```

```python
def plot_best_three_shooters(player_stats, n_players=10, min_3pa_per_game=2, min_games=20):
    """
    Mejores tiradores de 3 (porcentaje), filtrando por volumen mínimo.
    """
    df_plot = player_stats[
        (player_stats['games_played'] >= min_games) &
        (player_stats['three_att_per_game'] >= min_3pa_per_game) &
        (~player_stats['three_pct'].isna())
    ].copy()

    df_plot = df_plot.sort_values('three_pct', ascending=False).head(n_players)

    plt.figure(figsize=(10, 6))
    sns.barplot(
        data=df_plot,
        x='three_pct',
        y='fullName',
        order=df_plot.sort_values('three_pct', ascending=False)['fullName']
    )
    plt.title(f"Mejores tiradores de 3 puntos (Top {n_players})")
    plt.xlabel("Porcentaje de 3 (fracción, 0–1)")
    plt.ylabel("Jugador")
    plt.tight_layout()
    plt.show()
```

```python
def plot_best_free_throw_shooters(player_stats, n_players=10, min_fta_per_game=2, min_games=20):
    """
    Mejores tiradores de tiros libres, filtrando por volumen mínimo.
    """
    df_plot = player_stats[
        (player_stats['games_played'] >= min_games) &
        (player_stats['ft_att_per_game'] >= min_fta_per_game) &
        (~player_stats['ft_pct'].isna())
    ].copy()

    df_plot = df_plot.sort_values('ft_pct', ascending=False).head(n_players)

    plt.figure(figsize=(10, 6))
    sns.barplot(
        data=df_plot,
        x='ft_pct',
        y='fullName',
        order=df_plot.sort_values('ft_pct', ascending=False)['fullName']
    )
    plt.title(f"Mejores tiradores de tiros libres (Top {n_players})")
    plt.xlabel("Porcentaje de FT (fracción, 0–1)")
    plt.ylabel("Jugador")
    plt.tight_layout()
    plt.show()
```

```python
def plot_offensive_index(player_stats, n_players=10, min_minutes=300):
    """
    Índice ofensivo simple por 36 minutos:
    off_index = points + 1.5*assists + 0.7*rebounds_offensivos (todo per 20).
    """
    df_plot = player_stats[player_stats['total_minutes'] >= min_minutes].copy()

    # Rebotes ofensivos por 20
    eps = 1e-6
    den = np.where(df_plot['total_minutes'] > 0, df_plot['total_minutes'], eps)
    df_plot['reb_off_per_20'] = df_plot['total_reb_off'] / den * 20

    df_plot['off_index_per_20'] = (
        df_plot['points_per_20'] +
        1.5 * df_plot['assists_per_20'] +
        0.7 * df_plot['reb_off_per_20']
    )

    df_plot = df_plot.sort_values('off_index_per_20', ascending=False).head(n_players)

    plt.figure(figsize=(10, 6))
    sns.barplot(
        data=df_plot,
        x='off_index_per_20',
        y='fullName',
        order=df_plot.sort_values('off_index_per_20', ascending=False)['fullName']
    )
    plt.title(f"Índice ofensivo por 20 minutos (Top {n_players})")
    plt.xlabel("Índice ofensivo (adimensional)")
    plt.ylabel("Jugador")
    plt.tight_layout()
    plt.show()
```

```python
def plot_plus_minus(player_stats, kind='top', n_players=10, min_games=20):
    """
    Jugadores con mayor / menor plus-minus promedio.
    """
    df_plot = player_stats[player_stats['games_played'] >= min_games].copy()

    if kind == 'top':
        df_plot = df_plot.sort_values('plus_minus_avg', ascending=False).head(n_players)
        title_kind = "mayor impacto positivo"
        asc = False
    elif kind == 'bottom':
        df_plot = df_plot.sort_values('plus_minus_avg', ascending=True).head(n_players)
        title_kind = "mayor impacto negativo"
        asc = True
    else:
        raise ValueError("kind debe ser 'top' o 'bottom'.")

    plt.figure(figsize=(10, 6))
    sns.barplot(
        data=df_plot,
        x='plus_minus_avg',
        y='fullName',
        order=df_plot.sort_values('plus_minus_avg', ascending=asc)['fullName']
    )
    plt.title(f"Jugadores con {title_kind} (plus-minus promedio, min {min_games} juegos)")
    plt.xlabel("Plus-minus promedio")
    plt.ylabel("Jugador")
    plt.tight_layout()
    plt.show()
```

```python
def plot_players_points_by_year(df, kind='top', n_players=5):
    """
    Grafica la evolución de puntos por año para los jugadores con más o menos puntos totales.

    Parámetros
    ----------
    df : pandas.DataFrame
        DataFrame con columnas:
        'firstName', 'lastName', 'year', 'points'.
    kind : str
        'top'    -> jugadores con más puntos totales.
        'bottom' -> jugadores con menos puntos totales.
    n_players : int
        Número de jugadores a visualizar.
    """

    # Crear nombre completo
    df = df.copy()
    df['fullName'] = df['firstName'] + ' ' + df['lastName']

    # Puntos totales por jugador
    total_points = (
        df.groupby('fullName')['points']
          .sum()
          .reset_index(name='total_points')
    )

    # Elegir top o bottom
    if kind == 'top':
        selected_players = (
            total_points.sort_values('total_points', ascending=False)
                        .head(n_players)['fullName']
        )
        title_kind = "más anotadores"
    elif kind == 'bottom':
        selected_players = (
            total_points.sort_values('total_points', ascending=True)
                        .head(n_players)['fullName']
        )
        title_kind = "menos anotadores"
    else:
        raise ValueError("El parámetro 'kind' debe ser 'top' o 'bottom'.")

    # Filtrar solo esos jugadores
    df_sel = df[df['fullName'].isin(selected_players)].copy()

    # Asegurar orden por año
    df_sel = df_sel.sort_values('year')

    plt.figure(figsize=(15, 5))
    sns.lineplot(
        data=df_sel,
        x='year',
        y='points',
        hue='fullName',
        marker='o'
    )
    plt.title(f"Puntos promedio por partido a lo largo del tiempo\nJugadores {title_kind} (Top {n_players})")
    plt.xlabel("Año")
    plt.ylabel("Puntos")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```
